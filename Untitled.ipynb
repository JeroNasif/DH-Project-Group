```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_location = "properati.csv"
```

```python
data = pd.read_csv(data_location, sep=",")
data.shape
```

```python
data.head(20)
```

```python
#vemos los barrios

capital_federal_data = data[data.state_name == "Capital Federal"]
capital_federal_data.place_name.unique()

# definimos palermo palermo :  

# palermo soho
# palermo
# las cañitas
# palermo hollywood
# palermo chico
# palermo viejo
```

```python
#HACEMOS UNA LISTA PALERMO
lista_palermo = ['Palermo Soho', 'Palermo', 'Las Cañitas', 'Palermo Hollywood', 'Palermo Viejo', 'Palermo Chico' ]
```

```python
#realizamos bucle para que cada elemento de la lista sea un df

for i in range(0,len(lista_palermo)):
    lista_palermo[i]= data[data.place_name == lista_palermo[i]]

#concatenamos los df

df_palermo =pd.concat(lista_palermo)

print(df_palermo.place_name.value_counts())

print(df_palermo.shape)
```

```python
# Reseteo índices

df_palermo = df_palermo.reset_index()
df_palermo
```

```python
#Vemos qué hay en las columnas de precios

df_palermo_precios = df_palermo.loc[:,'price':'price_per_m2']

df_palermo_precios

print(df_palermo_precios[df_palermo_precios.currency == 'USD'])

print(df_palermo.currency.value_counts())

df_palermo.head(3)
```

```python
# Vemos cuántos nulos hay en cada columna.

print(df_palermo.isnull().sum())

df_palermo.isnull().sum()/df_palermo.shape[0]*100
```

```python
# Eliminar los registros que no tienen superficie cubierta

#df_palermo.surface_covered_in_m2.isnull().sum()

df_palermo = df_palermo.drop(df_palermo[df_palermo['surface_covered_in_m2'].isnull()==True].index)

df_palermo.shape

```

```python
#Buscamos registros con superficie cubierta mayor a superficie total

df_palermo_error_surface = df_palermo[(df_palermo.surface_total_in_m2 < df_palermo.surface_covered_in_m2)]

df_palermo_error_surface.loc[:,'surface_total_in_m2':'rooms']


# Eliminar filas que tienen superficie total < superficie cubierta 
# (otra opción sería tomar como válido el valor de superficie cubierta e imputarselo a ambos)


df_palermo = df_palermo.drop(df_palermo[(df_palermo.surface_total_in_m2 < df_palermo.surface_covered_in_m2) == True].index)


# df_palermo[(df_palermo.surface_total_in_m2 < df_palermo.surface_covered_in_m2)].index
# df_palermo[(df_palermo.surface_covered_in_m2 < df_palermo.surface_covered_in_m2) == True].index

df_palermo.shape
```

```python
df_palermo_sin_surface = df_palermo[df_palermo.surface_total_in_m2.isnull()]
df_palermo_sin_surface
```

```python
#Imputamos a superficie total el valor de superficie cubierta

# df_palermo.iat[120, 6] == 'NaN'

for row in range(0,len(df_palermo)):
    if df_palermo.iloc[row, df_palermo.columns.get_loc("surface_total_in_m2")] == 'nan':
        df_palermo.iloc[row, df_palermo.columns.get_loc("surface_total_in_m2")] = df_palermo.iloc[row, df_palermo.columns.get_loc("surface_covered_in_m2")].value

        
df_palermo.iloc[44, df_palermo.columns.get_loc("surface_total_in_m2")] = 5

df_palermo.iloc[4220, df_palermo.columns.get_loc("surface_total_in_m2")] #== 'nan'
```

```python
#borrar columnas:
    
lista_columnas_eliminar = ['index','Unnamed: 0', 'operation', 'place_with_parent_names', 'country_name', 'state_name', 'geonames_id',
                           'price', 'currency', 'price_aprox_local_currency', 'price_usd_per_m2', 'price_per_m2', 'expenses', 
                           'properati_url', 'image_thumbnail']

for columna in lista_columnas_eliminar:
    df_palermo.drop([columna], axis=1)

    
# unnamed?
# operation: todos tienen valor "sell"
# place_with_parent_names: redundante
# country_name
# state_name
# geonames_id redundante con nombre de barrio
# price: toma valores en la moneda que especifica la columna 'currency', es un campo heterogéneo
# currency
# price_aprox_local_currency: precio en ARS
# price_usd_per_m2: vamos a calcular un campo nuevo a partir de campos precio y superficie.
# price_per_m2: no entendemos a qué corresponde este valor
# expenses: mucho nan
# properati_url
# image_thumbnail
df_palermo = df_palermo.drop(lista_columnas_eliminar, axis=1)
df_palermo.shape
```

```python
#Hacemos una tabla dinámica para analizar por tipo de propiedad y cantidad de ambientes la variable superficie cubierta en metros cuadrados.



df_palermo.pivot_table('surface_covered_in_m2', ['property_type', 'rooms'],aggfunc=('count','sum','min','max','mean'))


```

```python
#Creamos dos columnas nuevas, una con el precio por m2 calculado con el precio total y la superficie total, y la otra con la superficie "descubierta" restandole la superficie cubierta a la total.

df_palermo['precio_m2_calc'] = df_palermo.price_aprox_usd / df_palermo.surface_total_in_m2

df_palermo['surface_uncovered_in_m2'] = df_palermo.surface_total_in_m2 - df_palermo.surface_covered_in_m2

print(df_palermo['surface_uncovered_in_m2'])

print(df_palermo.shape)

df_palermo.columns

df_palermo.head(2)
```

```python
# Comparo campo precio por metro cuadrado calculado con campo precio por metro cuadrado USD


serie_dif_pm2_usd = df_palermo.precio_m2_calc - df_palermo.price_usd_per_m2
print((serie_dif_pm2_usd > 1).value_counts())
print((serie_dif_pm2_usd < -1).value_counts())

```

```python
#La mayoría de los casos de nulos no coinciden entre precio y superficie, por lo que la cantidad de precio por m2 nulos es grande.
print('Nulos en campo price per m2: ' + str(df_palermo[df_palermo.price_usd_per_m2.isnull()].shape[0]))
print('Nulos en campo price: ' + str(df_palermo[df_palermo.price.isnull()].shape[0]))
print('Nulos en campo surface_total: ' + str(df_palermo[df_palermo.surface_total_in_m2.isnull()].shape[0]))
print('Nulos en campo price per m2 calc: ' + str(df_palermo[df_palermo.price_usd_per_m2.isnull()].shape[0]))
print('Registros con tanto price como surface_total nulos: ' + str(df_palermo[df_palermo.price.isnull() & df_palermo.surface_total_in_m2.isnull()].shape[0]))
```
